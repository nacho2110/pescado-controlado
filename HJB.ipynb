# Planteamiento de Hamilton-Jacobi-Bellman

Consideramos para el tiempo final $T$ la función valor
$$
\begin{align*}
V(t,x,y) &= \max_{E(\cdot)} \int_t^T e^{-\delta s} \pi_t ds\\
&= \max_{E(\cdot)} \int_t^T e^{-\delta s} \left[ (p_1 x(s) + p_2 y(s) - c) E(s) \right] ds\\
&= -\min_{E(\cdot)} \int_t^T -e^{-\delta s} \left[ (p_1 x(s) + p_2 y(s) - c) E(s) \right] ds
\end{align*}\\
$$
que representa el valor máximo a de las ganancias futuras desde el estado $(x,y)$ en el tiempo $t$ hasta el tiempo final $T$, bajo la dinámica dada por el sistema.



Con ello, $V(\cdot)$ satisface la ecuación de Hamilton-Jacobi-Bellman
$$
\begin{align*}
-\partial_t V(t,x,y) &= \max_{E \in [0,E_{max}]} \left\{ L(t,x,y,E) +
\partial_x V \cdot \left( rx(1-\frac{x}{K}) - axy - q_1xE \right) + 
\partial_y V \cdot \left( abxy - my - q_2yE \right)\\
\right\}\\
&= \max_{E \in [0,E_{max}]} \left\{ (p_1 x + p_2 y - c) E e^{-\delta t} +
\partial_x V \cdot \left( rx(1-\frac{x}{K}) - axy - q_1xE \right) + 
\partial_y V \cdot \left( abxy - my - q_2yE \right)\right\}\\
&= \partial_x V \left[rx(1-\frac{x}{K})- axy \right]
+ \partial_y V \left[abxy - my \right] + \max_{E \in [0,E_{max}]} {E \cdot \varphi(t,x,y)}\\ 
\end{align*}
$$
con $\varphi(t,x,y) = (p_1 x + p_2 y - c) e^{-\delta t} - q_1 x \partial_x V - q_2 y \partial_y V$.

Esto bajo la condición de borde
$$
V(T,x,y) = 0 \quad \text{para todo} (x,y)
$$
